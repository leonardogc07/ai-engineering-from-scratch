```python
import os
from dotenv import load_dotenv
load_dotenv()

# Hugging Face is 100% free (no credit card ever needed)
hf_token = os.getenv("HUGGINGFACE_API_KEY") or "hf_dummy"  # we'll use public endpoint for now

print("✅ Using free Hugging Face Inference (no quota issues)")

# Simple free call using the public inference API
import requests
response = requests.post(
    "https://api-inference.huggingface.co/models/google/gemma-2-2b-it",
    headers={"Authorization": f"Bearer {hf_token}"},
    json={"inputs": "What is a neural network in one sentence?"}
)

if response.status_code == 200:
    result = response.json()
    print("🚀 Successful free call:")
    print(result[0]["generated_text"] if isinstance(result, list) else result)
else:
    print("Status:", response.status_code)
    print(response.text[:300])
```
